# Scraping data of all matches

# ===============================================================

```python
from selenium import webdriver

website = 'https://www.adamchoi.co.uk/teamgoals/detailed'
path = "C:/Users/bahal/chromedriver/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get(website)

all_matches_button = driver.find_element_by_xpath("//label[@analytics-event = 'All matches']")
all_matches_button.click()

matches = driver.find_elements_by_tag_name('tr')

date = []
home_team = []
score = []
away_team = []

for match in matches:
    date.append(match.find_element_by_xpath('./td[1]').text)
    home_team.append(match.find_element_by_xpath('./td[2]').text)
    score.append(match.find_element_by_xpath('./td[3]').text)
    away_team.append(match.find_element_by_xpath('./td[4]').text)

# driver.quit()
```

```python
import pandas as pd

dt = {'Date':date,
      'Home Team':home_team,
      'Score':score,
      'Away Team':away_team}

df = pd.DataFrame(dt)

df
```

```python
df.to_csv('Data_of_adamchoi.csv', index = False)
```

# Scraping data of matches which was held in Spain (one country)

# ===============================================================

```python
import time
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd
```

```python
website = 'https://www.adamchoi.co.uk/teamgoals/detailed'
path = "C:/Users/bahal/chromedriver/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get(website)

all_matches_button = driver.find_element_by_xpath("//label[@analytics-event = 'All matches']")
all_matches_button.click()

dropdown = Select(driver.find_element_by_id('country'))
dropdown.select_by_visible_text('Spain')
                  
time.sleep(3)

matches_spain = driver.find_elements_by_tag_name('tr')

date_spain = []
home_team_spain = []
score_spain = []
away_team_spain = []

for match_spain in matches_spain:
    date_spain.append(match_spain.find_element_by_xpath('./td[1]').text)
    home_team_spain.append(match_spain.find_element_by_xpath('./td[2]').text)
    score_spain.append(match_spain.find_element_by_xpath('./td[3]').text)
    away_team_spain.append(match_spain.find_element_by_xpath('./td[4]').text)
                  
driver.quit()
                  
data_of_spain_matches = pd.DataFrame({'Date':date_spain,
                                      'Home Team':home_team_spain,
                                      'Score':score_spain,
                                      'Away Team':away_team_spain})
                  
data_of_spain_matches.to_csv('Data_of_Spain_matches.csv', index = False)

data_of_spain_matches
```
